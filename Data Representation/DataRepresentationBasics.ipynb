```python
sns.jointplot(x = "Customer_rating", y = "Prior_purchases",
              kind = "scatter", data = df)
plt.show()
```

```python
sns.jointplot(x = "Weight_in_gms", y = "Discount_offered",
              kind = "scatter", data = df)
plt.show()
```
